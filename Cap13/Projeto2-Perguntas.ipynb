## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
df_dsa_pn5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda', ascending = False)
df_dsa_pn5.head()
```

```python
def autopct_format(values): 
    def formatter(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return formatter
```

```python
plt.figure(figsize = (16, 8))


plt.pie(df_dsa_pn5['Valor_Venda'], 
        labels = df_dsa_pn5['Segmento'],
        autopct = autopct_format(df_dsa_pn5['Valor_Venda']),
        startangle = 90)


centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_pn5['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas por Segmento')
plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
df_dsa['Desconto'] = np.where(df_dsa['Valor_Venda'] > 1000, 0.15, 0.10)
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
df_dsa['Valor_Venda_Desconto'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])
```

```python
df_dsa.head()
```

```python
df_dsa_pn8_antes_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']
```

```python
df_dsa_pn8_depois_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda_Desconto']
```

```python
media_antes_desconto = df_dsa_pn8_antes_desconto.mean()
```

```python
media_depois_desconto = df_dsa_pn8_depois_desconto.mean()
```

```python
print('Média das vendas antes do desconto de 15% é igual a', round(media_antes_desconto, 2))
```

```python
print('Média das vendas depois do desconto de 15% é igual a', round(media_depois_desconto, 2))
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual a Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month
```

```python
df_dsa_pn9 = df_dsa.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
df_dsa_pn9
```

```python
anos = df_dsa_pn9.index.get_level_values(0)
meses = df_dsa_pn9.index.get_level_values(1)
segmentos = df_dsa_pn9.index.get_level_values(2)
```

```python
plt.figure(figsize = (16, 8))
sns.set()
fig1 = sns.relplot(kind = 'line',
                 data = df_dsa_pn9,
                 y = 'mean',
                 x = meses,
                 hue = segmentos,
                 col = anos,
                 col_wrap = 4)
plt.show()
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
df_dsa.head()
```

```python
df_dsa_pn10 = df_dsa.groupby(['Categoria', 
                              'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                    ascending = False).head(12)
```

```python
df_dsa_pn10 = df_dsa_pn10[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
df_dsa_pn10
```

```python
df_dsa_pn10_cat = df_dsa_pn10.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
df_dsa_pn10_cat
```

```python
cores_categorias = ['#5d00de',
                     '#0ee84f',
                     '#e80e27']
```

```python
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
fig, ax = plt.subplots(figsize = (20,12))

p1 = ax.pie(df_dsa_pn10_cat['Valor_Venda'],
           radius = 1,
           labels = df_dsa_pn10_cat['Categoria'],
           wedgeprops = dict(edgecolor = 'white'),
           colors = cores_categorias)

p2 = ax.pie(df_dsa_pn10['Valor_Venda'],
           radius = 0.9,
           labels = df_dsa_pn10['SubCategoria'],
           autopct = autopct_format(df_dsa_pn10['Valor_Venda']),
           colors = cores_subcategorias,
           labeldistance = 0.7,
           wedgeprops = dict(edgecolor = 'white'),
           pctdistance = 0.53,
           rotatelabels = True)

centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_pn10['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de Vendas por Categoria e Top 12 SubCategorias')
plt.show()
```
